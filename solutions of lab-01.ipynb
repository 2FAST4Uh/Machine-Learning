```python
import pandas as pd

# Load CSV file
df = pd.read_csv("insurance.csv")   # change filename if needed

# Calculate correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

# Choose target column
target = "charges"   # change if your target column name is different

print("Correlation values with", target)
print(corr[target])

# Check correlation type
positive = corr[target][corr[target] > 0].count()
negative = corr[target][corr[target] < 0].count()

print("\nSummary:")
if positive > negative:
    print("Dataset shows more POSITIVE correlation")
elif negative > positive:
    print("Dataset shows more NEGATIVE correlation")
else:
    print("Dataset has balanced or NO clear correlation")
```
